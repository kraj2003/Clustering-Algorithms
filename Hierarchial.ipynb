```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df.columns=iris.feature_names
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(df)
```

```python
X_scaled
```

```python
# apply pca
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca_scaled_data=pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled_data[:,0],pca_scaled_data[:,1],c=iris.target)
```

```python
# apply aglomatrive clustering
# construct a dendogram
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(8,5))
plt.title("Dendogram")
# create dendogram
sc.dendrogram(sc.linkage(pca_scaled_data,method='ward'))
plt.title("Dendogram")
plt.xlabel("Sample Index")
plt.ylabel("Euclideans Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_scaled_data)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled_data[:,0],pca_scaled_data[:,1],c=cluster.labels_)
```
